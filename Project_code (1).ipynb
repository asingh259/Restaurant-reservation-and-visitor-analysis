## 7. Clean Data <a id='clean_data'></a>

- Lets start the cleaning process by checking if there are rows full of NaN values.<br>
- We check the ID's for validating the data.<br>
- Then, columns are evaluated for datatypes and are adjusted.<br>
- Finally, a sequence of steps are executed to clean the remaining data.

#### 7.1 Checking for Nan rows: <a id='7.1'></a>

```python
def check_nan_rows(df_merged_final):
    ## Store all rows with NaN values
    nan_df = df_merged_final[pd.isnull(df_merged_final).any(axis=1)]
    print(' Number of rows that have at least ONE NaN value:', len(nan_df.index))
    #display(nan_df.head())

    ## List only rows that have NaN values (in all columns)
    nan_rows_df = nan_df[nan_df.isnull().all(1)]      
    nan_rows_count = len(nan_rows_df.index)
    print(' Number of rows completely filled with NaN values:', nan_rows_count)    
    #display(nan_rows_df.head())
    if (nan_rows_count != 0):
        print(' Oopsie! Found', nan_rows_count, 'NaN rows:')
        display(nan_rows_df.head())    
```

```python
check_nan_rows(df_merged_final)  
```

####  7.2 Checking for valid ID's: <a id='7.2'></a>

```python
def check_IDs(df_merged_final):
    # Count how many IDs are NaNs
    print(' NaN count on air_store_id:', df_merged_final.air_store_id.isnull().sum())
    print(' NaN count on hpg_store_id:', df_merged_final.hpg_store_id.isnull().sum())

    # Counting how many rows don't have a valid ID (i.e. air_store_id and hpg_store_id are both NaNs)
    null_ids_df = df_merged_final[df_merged_final.air_store_id.isnull() & df_merged_final.hpg_store_id.isnull()]
    null_ids_count = len(null_ids_df.index)
    print(' Number of rows without any IDs:', null_ids_count)
    
    if (null_ids_count != 0):
        display(null_ids_df.head())

    # Check how many IDs don't have any data at all (i.e. all columns are NaNs)
    max_nan_per_row = len(df_merged_final.columns) - 1    
    df_merged_final['NaN_count'] = df_merged_final.isnull().sum(axis=1)    
    invalid_ids_df = df_merged_final[df_merged_final['NaN_count'] == max_nan_per_row]    
    
    invalid_ids_count = len(invalid_ids_df.index)
    print(' Number of rows with IDs but NO DATA at all connected to them:', invalid_ids_count)
    
    if (invalid_ids_count != 0):
        print(' Found', invalid_ids_count, ' rows with IDs that have NO DATA at all')
        display(invalid_ids_df.head())
        
    del df_merged_final['NaN_count']
```

```python
check_IDs(df_merged_final)
```

#### 7.3 Checking for data types: <a id='7.3'></a>

```python
def get_data_types(df_merged_final):    
    col_names = df_merged_final.columns.tolist()
    data_types_df = pd.DataFrame(columns=col_names)    
    
    row = []
    for col in col_names:        
        row.append(str(df_merged_final[col].dtype))
    data_types_df.loc[0] = row
    return data_types_df
```

#### 7.4 Checking for corrupt data (negative values) :<a id='7.4'></a>

```python
#Checks numeric columns for negative numbers
def check_numeric(df_merged_final):     
    neg_count_df = pd.DataFrame(columns=['Negative values count'])
    
    #print(' Number of negative values found in numeric columns:')
    num_col_list = list(df_merged_final.select_dtypes(include=['int64', 'float64']).columns)    
    
    total_neg = 0
    for col_name in num_col_list:
        neg_count = df_merged_final[df_merged_final[col_name] < 0].shape[0]  # extract number of rows        
        neg_count_df.loc[col_name] = neg_count
        total_neg += neg_count
    
    return neg_count_df, total_neg
```

```python
neg_count_df, total_neg_count = check_numeric(df_merged_final)
display(neg_count_df)
```

#### 7.5 Checking for invalid data: <a id='7.5'></a>

```python
#Checks columns separately for NaN values
def check_invalid(df_merged_final): 
    nan_count_df = pd.DataFrame(columns=['NaN values count'])
    
    #print(' Number of NaN values found in each column:')
    col_list = list(df_merged_final.columns.tolist())    
    
    total_nan = 0
    for col_name in col_list:
        nan_count = df_merged_final[col_name].isnull().sum()        
        nan_count_df.loc[col_name] = nan_count        
        total_nan += nan_count
    
    return nan_count_df, total_nan
```

```python
#count how many NaNs are there
nan_count_df, total_nan_count = check_invalid(df_merged_final)
display(nan_count_df)
```

#### 7.6 Missing data visualization: <a id='7.6'></a>

```python
#plots a pie chart to visualize data corruption
def plot_inAccuracy_dis(total_values, total_neg_count, total_nan_count):
    neg_portion = total_neg_count / total_values * 100
    nan_portion = total_nan_count / total_values * 100
    ok_portion  =  (total_values - (neg_portion + nan_portion)) / total_values * 100

    # Data to plot
    labels = 'Negative values', 'NaN values', 'Proper values'
    sizes = [neg_portion, nan_portion, ok_portion]
    colors = ['red', 'orange', 'green']
    explode = (0.2, 0.2, 0)  # explode 1st and 2nd slice

    # Plot
    plt.subplots(figsize=(9,4))
    patches, texts = plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, startangle=330)
    plt.legend(patches, labels, loc="best")
    plt.axis('equal')
    plt.show()
    
    print('', round(nan_portion, 1), '% of the values are NaNs.')
```

```python
total_values = df_merged_final.shape[0] * df_merged_final.shape[1]
plot_inAccuracy_dis(total_values, total_neg_count, total_nan_count)
```

```python
def km_format(num):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.2f%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])
```

#### Plot the number of reservations (per week) in contrast with the number of visitations (per week).

```python
def plot_reservation_vs_visitors(df_merged_final):
    df_merged_final = df_merged_final.set_index('visit_date')
    reservations_made_df = df_merged_final.resample('W').apply({'visitors':'count'})
    visitors_df = df_merged_final.resample('W').apply({'visitors':'sum'})
    activity_df = pd.concat([reservations_made_df, visitors_df], join='inner', axis=1)
    activity_df.columns = ['reservations_made', 'visitors']
    fig, ax = plt.subplots(figsize=(15,5))

    # alternative method: use matplotlib's plot_date() and configure xticks at will
    plt.plot_date(activity_df.index.to_pydatetime(), activity_df, fmt='-')
    plt.title('Number of reservations made VS visitors (per week)')    
    ax = plt.gca()
    ax.xaxis.grid(True, which="major")
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%y'))
    d = activity_df.index.to_pydatetime()
    h_v = plt.fill_between(d, activity_df['visitors'], activity_df['reservations_made'], 
                           activity_df['visitors'] >= activity_df['reservations_made'],
                           facecolor='orange', alpha=0.5, interpolate=True)
    h_r = plt.fill_between(d, 0, activity_df['reservations_made'], 
                           facecolor='blue', alpha=0.5, interpolate=True)
    
    plt.legend(handles=[h_v, h_r], labels=["Visitors", "Reservations"], loc='upper left')
    ax.set_ylabel('Number of people')
    ax.margins(0.005, 0) # set margins to avoid "whitespace" while showing the first x-tick
    plt.show()
```

```python
plot_reservation_vs_visitors(df_merged_final)
```
